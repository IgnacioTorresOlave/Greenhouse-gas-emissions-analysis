```python
import matplotlib.pyplot as plt

selected_countries = ["France", "United States", "China", "Chile"]

plt.figure(figsize=(10,6))

col = "Per capita greenhouse gas emissions including land use"

for country in selected_countries:
    data = df[df['Entity'] == country]
    plt.plot(data['Year'], data[col], label=country)

plt.title("Per Capita GHG Emissions (Selected Countries)")
plt.xlabel("Year")
plt.ylabel("Emissions per capita")
plt.legend()
plt.grid(True)
plt.show()
```

# Key Insights

- Countries show different emission trends over time
- China shows strong growth
- France remains relatively stable
- USA remains high but declining
